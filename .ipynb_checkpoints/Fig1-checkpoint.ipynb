```python
# Vitualization of SPC

# folder = '/data0/cly/spectralGan/GAN-collections/results/unshift/lsun-church_outdoor/WGAN_GP_128_50_evaluate/*.png'
folder = '/data0/cly/spectralGan/GAN-collections/data/lsun/church_outdoor128/*.png'
number_iter = 5000

from torch.utils.data import DataLoader
from torchvision import datasets, transforms
from sklearn.decomposition import PCA
from scipy.interpolate import griddata
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import radialProfile, glob, cv2

spectrum_mat = np.zeros([number_iter, 128*128])
cnt = 0
spectrum_f = np.zeros([128, 128], dtype='float64')

for filename in glob.glob(folder):
    img = cv2.imread(filename, 0)
    f = np.fft.fft2(img)
    fshift = np.fft.fftshift(f)
    fshift += 1e-8

    magnitude_spectrum = 20*np.log(np.abs(fshift))
    spectrum_f += magnitude_spectrum
    spectrum_mat[cnt,:] = magnitude_spectrum.flatten()
    cnt+=1
    
    if cnt == number_iter: break

spectrum_mat -= np.mean(spectrum_mat) 
spectrum_mat /= np.std(spectrum_mat) 
pca = PCA(n_components=16, svd_solver='auto')
pca.fit(spectrum_mat)

fig, axes = plt.subplots(1, 8, figsize=(9, 4),
                         subplot_kw={'xticks':[], 'yticks':[]},
                         gridspec_kw=dict(hspace=0.1, wspace=0.1))
for i, ax in enumerate(axes.flat):
    ax.imshow(pca.components_[i].reshape(128, 128), cmap='gray')
```

```python
# PSD Vitualization

from torch.utils.data import DataLoader
from torchvision import datasets, transforms

import radialProfile
import glob
import cv2
from scipy.interpolate import griddata
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

class DiskImageDataset():
    def __init__(self, img_paths, map_fn=None):
        self.img_paths = img_paths
        self.map_fn = map_fn
        
    def pil_loader(self, path):
        with open(path, 'rb') as f:
            return Image.open(f).convert('RGB')
        
    def __getitem__(self, i):
        if self.map_fn is not None:
            return self.map_fn(self.pil_loader(self.img_paths[i]))
        else:
            return self.pil_loader(self.img_paths[i])
    
    def __len__(self):
        return len(self.img_paths)

class church_draw():
    def __init__(self):
        self.epsilon = 1e-8
        self.number_iter = 1000
        self.y = []
        self.error = []
        self.rgb_weights = [0.2989, 0.5870, 0.1140]
        self.input_size = 128
        self.N = 88
        self.real_data = DiskImageDataset(glob.glob('data/lsun/church_outdoor' + str(self.input_size) + '/*.png'), map_fn=None)

    def get_psd1D(self, data=None, folder=None):
        psd1D_total = np.zeros([self.number_iter, self.N])
        psd1D_org_mean = np.zeros(self.N)
        psd1D_org_std = np.zeros(self.N)
        cnt = 0
        
        if data is not None:
            for iter, x_ in enumerate(data):
                x_ = np.dot(np.asarray(x_), self.rgb_weights)
                f = np.fft.fft2(x_)
                fshift = np.fft.fftshift(f)
                fshift += self.epsilon

                magnitude_spectrum = 20*np.log(np.abs(fshift))
                # Calculate the azimuthally averaged 1D power spectrum
                psd1D = radialProfile.azimuthalAverage(magnitude_spectrum)
                # Calculate the azimuthally averaged 1D power spectrum
                points = np.linspace(0, self.N, num=psd1D.size) # coordinates of a
                xi = np.linspace(0, self.N, num=self.N) # coordinates for interpolation
                interpolated = griddata(points, psd1D, xi, method='cubic')
                interpolated = (interpolated-np.min(interpolated))/(np.max(interpolated)-np.min(interpolated))
                psd1D_total[cnt,:] = interpolated  
                
                cnt+=1
                
                if cnt == self.number_iter:
                    break

        elif folder is not None:
            for filename in glob.glob(folder):
                img = cv2.imread(filename, 0)
                f = np.fft.fft2(img)
                fshift = np.fft.fftshift(f)
                fshift += self.epsilon

                magnitude_spectrum = 20*np.log(np.abs(fshift))

                # Calculate the azimuthally averaged 1D power spectrum
                psd1D = radialProfile.azimuthalAverage(magnitude_spectrum)
                psd1D = (psd1D-np.min(psd1D))/(np.max(psd1D)-np.min(psd1D))
                psd1D_total[cnt,:] = psd1D  

                cnt+=1
                
                if cnt == self.number_iter:
                    break

        for x in range(self.N):
            psd1D_org_mean[x] = np.mean(psd1D_total[:,x])
            psd1D_org_std[x] = np.std(psd1D_total[:,x])

        return psd1D_org_mean, psd1D_org_std


    def set_stat_main(self, real = True):

        if(real):
            real_psd1D_mean, real_psd1D_std =  self.get_psd1D(data=self.real_data)
            self.label_list = ['Real']
            self.y.append(real_psd1D_mean); self.error.append(real_psd1D_std)
        else:
            generated_psd1D_mean, generated_psd1D_std =  self.get_psd1D(folder="./results/unshift/lsun-church_outdoor/WGAN_GP_128_50_evaluate/*.png")
            self.label_list = ['WGAN-GP']
            self.y.append(generated_psd1D_mean); self.error.append(generated_psd1D_std)


    def draw(self):
        x = np.arange(0, self.N, 1)
        fig, ax = plt.subplots(figsize=(18, 12))
        
        for index, label in enumerate(self.label_list):
            ax.plot(x, self.y[index], alpha=0.5, label=label, linewidth=8.0)
            # ax.fill_between(x, self.y[i] - self.error[i], self.y[i] + self.error[i], alpha=0.2)
        
        plt.xlabel('Frequency', fontsize=35)
        plt.ylabel('Power Spectrum', fontsize=35)
        plt.tick_params(axis='x', labelsize=35)
        plt.tick_params(axis='y', labelsize=35)

        ax.legend(loc='best', prop={'size': 40})


plt.style.use("seaborn-paper")
```

```python
church_128x128_real = church_draw()
church_128x128_real.set_stat_main()
church_128x128_real.draw()

church_128x128_syn = church_draw()
church_128x128_syn.set_stat_main(real=False)
church_128x128_syn.draw()
```
