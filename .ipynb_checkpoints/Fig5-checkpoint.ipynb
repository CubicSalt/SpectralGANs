```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

r = [1, 2, 4, 8, 12, 16]

DCGAN_FID = [140.1881305, 141.3962259, 130.5007665, 127.1346926, 95.23025975, 80.49529035]
DRAGAN_FID = [20.92474219, 23.06318508, 25.39486327, 32.08204726, 24.77024513, 16.67151429]
LSGAN_FID = [50.86687579, 52.04921365, 58.15745092, 61.13505537, 38.1866764, 28.15849799]
WGAN_GP_FID = [59.52696481, 65.17018411, 81.02907845, 92.02258209, 51.74381205, 33.58451523]

DCGAN_FID_Deconv = [48.98816703, 45.83160301, 43.11147783, 55.14558303, 42.4907701, 26.90007995]
DRAGAN_FID_Deconv = [24.42169832, 26.32693041, 25.4853996, 29.03803376, 23.62965893, 17.1165848]
LSGAN_FID_Deconv = [44.10120451, 43.90523763, 43.87977434, 43.29209082, 31.8717472, 21.82547025]
WGAN_GP_FID_Deconv = [38.34241031, 40.46925355, 46.07415308, 57.64779393, 41.92360178, 30.06391523]

def draw(legend=True, y_axis=True):
    fig, ax = plt.subplots(figsize=(18, 12))

    ax.plot(r, DCGAN_FID, alpha=0.5, linestyle='--', label='Upsample DCGAN', linewidth=4.0)
    ax.plot(r, DRAGAN_FID, alpha=0.5, linestyle='--', label='Upsample DRAGAN', linewidth=4.0)
    ax.plot(r, LSGAN_FID, alpha=0.5, linestyle='--', label='Upsample LSGAN', linewidth=4.0)
    ax.plot(r, WGAN_GP_FID, alpha=0.5, linestyle='--', label='Upsample WGAN-GP', linewidth=4.0)
    
    ax.plot(r, DCGAN_FID_Deconv, alpha=0.5, label=  'Deconv DCGAN', linewidth=4.0)
    ax.plot(r, DRAGAN_FID_Deconv, alpha=0.5, label= 'Deconv DRAGAN', linewidth=4.0)
    ax.plot(r, LSGAN_FID_Deconv, alpha=0.5, label=  'Deconv LSGAN', linewidth=4.0)
    ax.plot(r, WGAN_GP_FID_Deconv, alpha=0.5, label='Deconv WGAN-GP', linewidth=4.0)
    

#     plt.title(, fontsize=35)
    plt.xlabel('Radius', fontsize=35)
    plt.ylim(0, 170)
    plt.ylabel('FID', fontsize=35)
    plt.tick_params(axis='x', labelsize=35)
    plt.tick_params(axis='y', labelsize=35)
    ax.legend(loc='best', prop={'size': 22})
    plt.style.use("seaborn-paper")
```

```python
draw()
```
