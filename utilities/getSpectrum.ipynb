```python
import matplotlib.pyplot as plt
import matplotlib.image
import numpy as np
import scipy.misc

import warnings
warnings.filterwarnings("ignore")

def getSpectrum(filepath):
    pic = matplotlib.image.imread(filepath)
    spec = np.array(pic)
    spec = 0.299 * spec[:, :, 0] + 0.587 * spec[:, :, 1] + 0.114 * spec[:, :, 2]
    spec = np.fft.fft2(spec).real
    spec = np.fft.fftshift(spec)
    spec += 1e-8
    spec = 20*np.log(np.abs(spec))

    plt.subplot(1,2,1)
    plt.xticks([])  
    plt.yticks([])
    plt.imshow(pic, cmap = 'gray')
    plt.subplot(1,2,2)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(spec, cmap = 'gray')
    scipy.misc.imsave(filepath[:-4] + '_Spectrum.jpg', spec)
```

```python
getSpectrum("../PDCAT/Inversion.jpg")
```

```python
getSpectrum("../PDCAT/Real.jpg")
```
